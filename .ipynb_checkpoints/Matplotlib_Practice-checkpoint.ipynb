```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot([1,2,3,4])
plt.ylabel("Some Numbers")
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
```

```python
plt.plot([1,2,3,4], [1,4,9,16], 'ro')
plt.axis((0,6,0,20))
plt.show()
```

```python
t = np.arange(0. , 5., 0.2)

plt.plot(t,t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
# Plotting with data
data = {'a' : np.arange(50),
       'c' : np.random.randint(0,50,50),
       'd' : np.random.randn(50)}
data['b'] = data['a'] + 10*np.random.randn(50)
data['d'] = np.abs(data['d']) *100

plt.scatter('a','b',c = 'c', s = 'd', data = data)
plt.xlabel("Entry A")
plt.ylabel("Entry B")
plt.show()
```

```python
x = np.linspace(0.0, 0.4*np.pi, 1000)
plt.plot(x, np.sin(x), 'r-')
```

```python
# Generating and Plotting a random distribution
mu, sigma = 0, 1 
x = mu + sigma * np.random.randn(10000)

plt.hist(x, 50, facecolor = 'blue')
```

```python
a = np.arange(0.0, 5.0, 0.01)
x = np.cos(2*np.pi*a)
line, = plt.plot(a, x, lw = 2)
plt.annotate('local max', xy=(2,1), xytext = (3, 1.5), arrowprops = dict(facecolor = 'green', shrink = 0.05),)
plt.show()
```
