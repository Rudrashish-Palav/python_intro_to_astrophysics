```python
# Left Endpoint Method
import numpy as np
dx = 0.01
x = np.arange(0.0, 2.0, dx)
y = x**2
Integral_Value = np.sum(y*dx)
print('Numerical estimate of integral is %8.8g. Analytical solution is %8.8g.'%(Integral_Value,1.0/3.0*2.0**3))
```

```python
# With midppoint method

import numpy
dx = 0.01
x = numpy.arange(0.0,2.0,dx)
x += dx/2
y = x**2
IntegralValue = numpy.sum(y*dx)
print('Numerical estimate of integral is %8.8g. Analytical solution is %8.8g.'%(IntegralValue,1.0/3.0*2.0**3))

```

```python
# Scipy for integration
import numpy
import scipy.integrate
dx = 0.01
x = numpy.arange(0.0,2.0,dx)
y = x**2
IntegralValue = scipy.integrate.simps(y,dx=dx)
print('With Simpsons method we get %8.8g. Analytical solution is %8.8g.'%(IntegralValue,1.0/3.0*2.0**3))
```

```python
# Trapezoidal Rule 
import numpy
import scipy.integrate
dx = 0.01
x = numpy.arange(0.0,2.0,dx)
y = x**2
IntegralValue = scipy.integrate.trapz(y,dx=dx)
print('With Simpsons method we get %8.8g. Analytical solution is %8.8g.'%(IntegralValue,1.0/3.0*2.0**3))
```

```python
# Integration of a function
import numpy
import scipy.integrate
def f(t):
    return t**2
Result = scipy.integrate.romberg(f,0.0,2.0)
print('With Rombergs method for a function we get %18.18g'%Result)
Result = scipy.integrate.quad(f,0.0,2.0)
print('With the quad method for a function we get %18.18g +/− %5.5g'%(Result[0],Result[1]))
```

```python
import numpy as np
dx = 0.1
x = np.arange(0.0,2.0,dx)
x += dx/2
y = x**3
IntegralValue = np.sum(y*dx)
print('Numerical estimate of integral is %8.8g.'%(IntegralValue))

x = np.append(x,2.0)#append endpoint to plot
y = np.append(y,y[-1])#append endpoint to plot
plt.fill_between(x,y,color='red',step='mid',label='Estimate')
plt.title('Midpoint estimate for $dx=0.1$')
plt.legend(loc=2)
plt.xlabel('x')
plt.ylabel('y')
plt.tight_layout()

plt.show()
```

### Compute all using Simpson's method

```python
import scipy.integrate


dx = 0.01
x = numpy.arange(0.0,2.0,dx)
y = x**3


#Simpson integration (https://docs.scipy.org/doc/scipy/reference/generated/scipy.integrate.simps.html#scipy.integrate.simps)
Result = scipy.integrate.simps(y,dx=dx)
print('With Simpsons method for data we get %18.18g'%Result)
```
