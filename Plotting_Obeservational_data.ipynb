```python
absorbtion = {}
absorbtion['S2'] = 1253.8110
absorbtion['Ni2'] = 1370.1320
absorbtion['Si2'] = 1526.7066
absorbtion['Fe2'] = 1608.4511
absorbtion['Si'] =  1808.0126 
absorbtion['Cr'] = 2056.2540
absorbtion['Zn'] = 2062.6640
```

```python
import matplotlib.pyplot as plt

# Number of subplots based on the length of absorbtion keys
num_plots = len(absorbtion.keys())

# Create subplots with one row and num_plots columns
fig, axes = plt.subplots(7,1, figsize=(10, 20))

# Ensure axes is iterable even if num_plots = 1
if num_plots == 1:
    axes = [axes]

# Loop over absorption keys and corresponding axes
for idx, i in enumerate(absorbtion.keys()):
    ax = axes[idx]  # Select the current subplot axis
    
    # Plot the data
    ax.plot(NIR[:, 0], NIR[:, 1], 'r-', lw=2, label='NIR')
    ax.plot(VIS[:, 0], VIS[:, 1], 'b-', lw=2, label='VIS')
    ax.plot(Continuum[:, 0], Continuum[:, 1], color='black', ls='-', lw=2, label='Continuum')
    ax.plot(VIS_model[:, 0], VIS_model[:, 1], color='grey', ls='-', lw=2, label='VIS_model')
    ax.plot(NIR_model[:, 0], NIR_model[:, 1], color='grey', ls='-', lw=2, label='NIR_model')
    
    # Set axis limits and labels
    ax.set_ylim([-0.2e-16, 1e-16])
    ax.set_xlim((absorbtion[i] * (1 + 4.991) - 10, absorbtion[i] * (1 + 4.991) + 10))
    ax.set_title(f"Absorption at z = 4.991, key: {i}")
    
    # Add legend
    ax.legend(loc="upper left")
plt.tight_layout()
plt.show()
```
